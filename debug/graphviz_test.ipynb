```python
from graphviz import Digraph
```

```python
dot = Digraph(format='svg', graph_attr={'rankdir': 'LR'}, name='round-table', comment='The Round Table')
```

```python
dot.node(name='A', label='King Arthur')
dot.node(name='B', label='Sir Bedevere the Wise')
dot.node(name='L', label='Sir Lancelot the Brave')
```

```python
dot.edges(['AB', 'AL'])
dot.edge('B', 'L', constraint='False')
```

```python
print(dot.source)
```

```python
dot.render(directory='doctest-output').replace('\\', '/')
```
