```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-20, 5, 100)
y = np.linspace(-5, 20, 100)
X, Y = np.meshgrid(x, y)
```

```python
def f(x: np.ndarray):
    "The function to be minimized"
    return x[0]**2 + x[1]**2
```

```python
def grad_f(x: np.ndarray):
    "The gradient of the function to be minimized"
    return np.array([2*x[0], 2*x[1]])
```

```python
lr = 0.1
epochs = 20
x0 = np.array([-15, 15])

grads = []
xs = [x0.copy()]
for i in range(epochs):
    grad = grad_f(x0)
    grads.append(grad)
    
    x0 = x0 - lr * grad
    xs.append(x0.copy())

grads = np.array(grads)
xs = np.array(xs)
```

```python
plt.figure(figsize=(10, 10))
ax = plt.axes(projection='3d')
ax.grid(False)
plt.title(r"Gradient descent on $f(x)=x_1^2+x_2^2$")

ax.plot_surface(X, Y, f(np.array([X, Y])), cmap='Blues',alpha=0.6)
ax.plot(xs[:, 0], xs[:, 1], f([*xs.T]), color='b', marker='x', markersize=10, linewidth=1.5, label='Gradient descent')
ax.scatter(0, 0, 0, color='r', label='Minima')

ax.set_xlabel(r'$x_1$')
ax.set_ylabel(r'$x_2$')
ax.set_zlabel(r'$f(x)$')
plt.legend(loc='center left')
plt.show()
```
